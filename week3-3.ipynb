Get data that in the table of postal codes and to transform the data into a pandas dataframe

```python
import pandas as pd

dataF = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)[0]
dataF
```

Ignore cells with a borough that is Not assigned

```python
dataF = dataF.drop(dataF[(dataF.Borough == 'Not assigned')].index)
dataF.reset_index(drop=True) 
```

Combine Neighbourhood for duplicate PostCode

```python
dataF['Neighbourhood']=dataF[['Postcode','Borough','Neighbourhood']].groupby(['Postcode','Borough'])['Neighbourhood'].transform(lambda x: ','.join(x))
dataF = dataF.drop_duplicates().reset_index(drop=True)
dataF
```

Replace Not assigned neighborhood by the Borough value

```python
dataF['Neighbourhood'] = dataF['Neighbourhood'].replace('Not assigned', dataF['Borough'])
dataF
```

```python
dataF.shape
```

```python
dataFCoordinate = pd.read_csv('https://cocl.us/Geospatial_data')
dataFCoordinate
```

```python
dataFMergered = dataF.merge(dataFCoordinate, left_on='Postcode', right_on='Postal Code', how='left')
```

```python
dataFMergered = dataFMergered.drop(['Postal Code'], axis=1)
```

Get only Boroughs that contain the word Toronto

```python
dataFToronto = dataFMergered[dataFMergered['Borough'].str.contains("Toronto")]
```

```python
from geopy.geocoders import Nominatim
import folium
```

Use geopy library to get the latitude and longitude values of Toronto.

```python
address = 'Toronto, ON'

geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

Create a map of Toronto with neighborhoods superimposed on top.

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(dataFToronto['Latitude'], dataFToronto['Longitude'], dataFToronto['Borough'], dataFToronto['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
